```python
import numpy as np
from ROOT import TH1D, TCanvas
```

```python
file1 = './Labs/muon_271119/Run3/3052B-01_1_1_113001_2010_191127-175401.dat'
```

```python
dt = np.genfromtxt(file1, delimiter='\n',usecols=[0],skip_header=13)
```

```python
for i in range(1):
    event = dt[1000*i:1000*(i+1)]
```

```python
hist = TH1D("hist","",1000,0,1000)
c = TCanvas()
for i in range(event.size):
    hist.SetBinContent(i+1,event[i])
hist.Draw()
c.Draw()
```
